```python
# Copying Lists in Python

# Referencing vs Copying
# ----------------------
# a = [1, 2, 3]
# b = a          -> Referencing (b just points to the same object as a)
# b = a.copy()   -> Shallow Copy (top-level copy; nested lists still share references)
# b = copy.deepcopy(a) -> Deep Copy (fully independent copy, including nested lists)

# Example 1: Referencing (no copy)
a = [1, 2, 3]
b = a
a[0] = 4
b[2] = 9
print(a, b)  # Output: [4, 2, 9] [4, 2, 9]

# Example 2: Shallow Copy
a = [1, [1, 3]]
b = a.copy()
a[0] = 4
a[1][1] = 9
print(a, b)  # Output: [4, [1, 9]] [1, [1, 9]]

# Explanation:
# Shallow copy only copies the outer list.
# The inner list is still a shared reference between a and b.

# Example 3: Deep Copy
import copy

a = [1, 2, [4]]
b = copy.deepcopy(a)
a[0] = 7
b[2][0] = 9
print(a, b)  # Output: [7, 2, [4]] [1, 2, [9]]
```

| Use `**kwargs` | When you don’t know how many keyword arguments will be passed    |
